```python
import pandas as pd
import scipy.stats as stats
from statsmodels.stats.multitest import multipletests
from itertools import combinations


# Create a contingency table
contingency_table = pd.crosstab(df1["Age_group"], df1["causes_of_kd"])

# Perform Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print results
print(f"Chi-square statistic: {chi2}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p}")

# Multiple comparisons: Pairwise chi-square tests with Bonferroni correction
age_groups = df1["Age_group"].unique()
cause_groups = df1["causes_of_kd"].unique()
pairwise_tests = []

for (age1, age2) in combinations(age_groups, 2):
    subset = df1[df1["Age_group"].isin([age1, age2])]
    cont_table = pd.crosstab(subset["Age_group"], subset["causes_of_kd"])
    
    if cont_table.shape[1] > 1:  # Ensure valid contingency table
        chi2_pair, p_pair, _, _ = stats.chi2_contingency(cont_table)
        pairwise_tests.append((age1, age2, p_pair))

# Adjust p-values for multiple comparisons
if pairwise_tests:
    comparisons_df = pd.DataFrame(pairwise_tests, columns=["Age_Group_1", "Age_Group_2", "P_Value"])
    comparisons_df["Adjusted_P_Value"] = multipletests(comparisons_df["P_Value"], method='bonferroni')[1]
    
    print("\nPairwise comparisons with Bonferroni correction:")
    print(comparisons_df)

```
